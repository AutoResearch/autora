# Notebooks

You can use [Jupyter Notebooks](https://jupyter.org) as part of the documentation for your contribution.

## Including Images
When using images, ensure they are "attached" to cells, rather than being imported from a path.
The image should appear like this in the markdown cell:

```markdown

```

To attach an image, you can open the notebook in the Jupyter notebook browser.

For example, to open the `docs/contribute/Notebook.ipynb` file, you can run:
```shell
jupyter notebook "docs/contribute/Notebook.ipynb"
```
